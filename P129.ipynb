```python
import pandas as pd

bright_stars_df = pd.read_csv("https://drive.google.com/uc?id=1NtqbM6tKjeYYaMxQN2kKjsVE_94ahLEF") #https://drive.google.com/file/d/1NtqbM6tKjeYYaMxQN2kKjsVE_94ahLEF/view?usp=sharing

brown_dwarf_df = pd.read_csv("https://drive.google.com/uc?id=1I2iRktqnG3zUngtpv2jsME5P_kIbkqtL") #https://drive.google.com/file/d/1I2iRktqnG3zUngtpv2jsME5P_kIbkqtL/view?usp=drive_link
brown_dwarf_df.head(10)
```

```python
brown_dwarf_df = brown_dwarf_df.dropna(subset=["Distance" , "Mass" , "Radius"])
brown_dwarf_df.head(10)
```

```python
brown_dwarf_df['Distance'] = brown_dwarf_df["Distance"].astype(float)
brown_dwarf_df['Mass'] = brown_dwarf_df["Mass"].astype(float)
brown_dwarf_df['Radius'] = brown_dwarf_df["Radius"].astype(float)

print(brown_dwarf_df.dtypes)
```

```python
brown_dwarf_df["Radius"] = brown_dwarf_df["Radius"] * 0.102763
brown_dwarf_df["Mass"] = brown_dwarf_df["Mass"] * 0.000954588

brown_dwarf_df.head(10)
```
